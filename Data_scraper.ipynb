```python
import nest_asyncio
import asyncio
import json
from twikit import Client
import logging
import pandas as pd
import os
from datetime import datetime


# Initialiserer AsyncIO til at køre i en Jupyter- eller interaktiv session
nest_asyncio.apply()


# Opsæt logging
logging.basicConfig(
    level=logging.INFO, 
    format='%(asctime)s - %(levelname)s: %(message)s'
)
logger = logging.getLogger(__name__)


# Cookies-fil
COOKIES_FILE = 'X_cookies.json'


# Funktion til at formatere cookies korrekt til httpx
def format_cookies_for_httpx(cookies):
    return {cookie['name']: cookie['value'] for cookie in cookies}


async def extract_tweet_data(tweets):
    """
    Ekstraherer tweet-data og opretter et Pandas DataFrame
    
    Args:
        tweets (list): Liste af tweet-objekter fra Twikit
    
    Returns:
        pandas.DataFrame: DataFrame med tweet-information
    """
    tweet_data = []
    
    for tweet in tweets:
        tweet_info = {
            'Tweet_id': tweet.id,
            'User_name': tweet.user.name,
            'Tweet_text': tweet.full_text,
            'Language': tweet.lang,
            'Hashtags': tweet.hashtags,
            'Created_at': tweet.created_at_datetime
            
        }
        tweet_data.append(tweet_info)
    
    return pd.DataFrame(tweet_data)

async def main():
    try:
        # Initialisér klienten
        client = Client(language='en-US')
        logger.info("Klient initialiseret.")
        
        # Indlæs cookies fra fil
        try:
            with open(COOKIES_FILE, 'r', encoding='utf-8') as f:
                cookies = json.load(f)
                
            # Konverter cookies til korrekt format
            formatted_cookies = format_cookies_for_httpx(cookies)
            client.set_cookies(formatted_cookies)  # Anvend de formaterede cookies
            logger.info("Cookies indlæst og appliceret.")
            
        except FileNotFoundError:
            logger.error(f"Cookies-fil '{COOKIES_FILE}' ikke fundet. Sørg for at gemme cookies først.")
            return None
        except Exception as e:
            logger.error(f"Fejl ved indlæsning af cookies: {e}")
            return None
        
        # Søgeord og parametre
        keywords = ['Tesla', 'Tesla Model Y', 'Tesla Model 3', 'CyberTruck']
        tweets_per_keyword = 60  # Ønsket antal tweets pr. keyword
        
        # Samlet DataFrame til at gemme alle tweets
        all_tweets_df = pd.DataFrame()
        
        # Søg efter tweets for hvert nøgleord
        for keyword in keywords:
            try:
                logger.info(f"Søger tweets for nøgleord: {keyword}")
                
                # Første søgning
                tweets = await client.search_tweet(keyword, 'Top', 20) # Do not increase from 20 
                
                # Ekstrahér første sæt tweets
                keyword_tweets_df = await extract_tweet_data(tweets)
                keyword_tweets_df['Search_Keyword'] = keyword
                all_tweets_df = pd.concat([all_tweets_df, keyword_tweets_df], ignore_index=True)
                
                # Hent flere tweets indtil vi når ønsket antal
                current_tweet_count = len(tweets)
                
                while current_tweet_count < tweets_per_keyword:
                    try:
                        # Hent næste sæt tweets
                        more_tweets = await tweets.next()
                        time.sleep(0.2)
                        
                        # Ekstrahér og tilføj nye tweets
                        more_tweets_df = await extract_tweet_data(more_tweets)
                        more_tweets_df['Search_Keyword'] = keyword
                        all_tweets_df = pd.concat([all_tweets_df, more_tweets_df], ignore_index=True)
                        
                        # Opdater tweet tæller
                        current_tweet_count += len(more_tweets)
                        
                        logger.info(f"Hentet yderligere tweets for '{keyword}'. Total: {current_tweet_count}")
                        
                        # Stop hvis vi har nået eller overskredet ønsket antal
                        if current_tweet_count >= tweets_per_keyword:
                            break
                    
                    except Exception as more_tweets_error:
                        logger.warning(f"Kunne ikke hente flere tweets for '{keyword}': {more_tweets_error}")
                        break
                
                logger.info(f"Fandt i alt {current_tweet_count} tweets for '{keyword}'")
            
            except Exception as search_error:
                logger.error(f"Fejl ved søgning af tweets for '{keyword}': {search_error}")
        
        # Fjern duplikater
        all_tweets_df.drop_duplicates(subset=['Tweet_id'], inplace=True)
        
        # Begræns til ønsket antal tweets
        all_tweets_df = all_tweets_df.head(tweets_per_keyword * len(keywords))
        
        # Opret mappe til output hvis den ikke eksisterer
        os.makedirs('output', exist_ok=True)
        
        # Gem DataFrame som CSV med tidsstempel
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        output_file = f'output/twitter_tweets_{timestamp}.csv'
        all_tweets_df.to_csv(output_file, index=False, encoding='utf-8')
        logger.info(f"Tweets gemt i {output_file}")
        
        # Returner DataFrame
        return all_tweets_df
    
    except Exception as e:
        logger.error(f"Overordnet fejl: {e}")
        return None
    
# Kør den asynkrone funktion
df = asyncio.run(main())


# Vis DataFrame (hvis den er blevet oprettet)
if df is not None:
    print("\nTweet DataFrame:")
    print(df.head())
    print(f"\nSamlet antal tweets: {len(df)}")
```
